```python
# X_0=3. dXt = -alpha x Xtdt + sig x dZt
# Euler process (sampling from Diffusion)

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
M = 100 
T = 1.0
dt = T/M

X_t = np.zeros(M+1)

X_t[0] = 0.5
alpha = 10
sigma = 2

# Euler Discretization 방식으로 simulation
for idx in range(M):
    dZ = np.random.randn()*np.sqrt(dt) # BM
    X_t[idx+1] = X_t[idx] - alpha*X_t[idx]*dt + sigma*dZ # 약간씩 error 포함됨
    
plt.plot(X_t)
plt.show()
```

```python
# Square-root process

import numpy as np
import matplotlib.pyplot as plt

# discretization 
np.random.seed(0)
M = 252 # daily sampling
T = 1.0
dt = T/M

X_t = np.zeros(M+1)

X_0 = 3
alpha = 1
sigma = 2
beta = 2 # mean-reverting level

# Euler Discretization 방식으로 simulation
X_t[0] = X_0

for idx in range(M):
    dZ = np.random.randn()*np.sqrt(dt) # BM
    X_t[idx+1] = X_t[idx] + alpha*(beta-X_t[idx])*dt + sigma*dZ # 약간씩 error 포함됨
    
plt.plot(X_t)
plt.show()
```
