```python
# 1. finding an efficient frontier based on two stocks by simulation
# (a) draw the efficient frontier with "short-selling constraint" 
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) 
N = int(1e4) # 10,000
mu_vector = np.array([0.15, 0.25]) 
sigma_vector = np.array([0.1, 0.2]) 
corr = 0.2

w1 = np.random.rand(N) # 공매도 제한 : 0과 1사이의 Uniform한 10,000개의 난수 
w2 = 1-w1 # weight의 합은 1

mu_pf = w1*mu_vector[0] + w2*mu_vector[1] 
var_pf = (w1**2)*(sigma_vector[0]**2) + (w2**2)*(sigma_vector[1]**2) + 2*w1*w2*corr*sigma_vector[0]*sigma_vector[1] # (ax+by)^2 포맷
std_pf = np.sqrt(var_pf)

plt.plot(std_pf, mu_pf, '.') # '.' 옵션을 주어 선으로만 표현
plt.show()
```

```python
# 1-(b) draw the efficient frontier without "short-selling constraint" 
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) 
N = int(1e4)
mu_vector = np.array([0.15, 0.25]) 
sigma_vector = np.array([0.1, 0.2]) 
corr = 0.2

w1 = np.random.randn(N) # 공매도 허용 : w1은 어떤 값도 취할 수 있음 (cf. randn)
w2 = 1 - w1

mu_pf = w1*mu_vector[0] + w2*mu_vector[1] 
var_pf = (w1**2)*(sigma_vector[0]**2) + (w2**2)*(sigma_vector[1]**2) + 2*w1*w2*corr*sigma_vector[0]*sigma_vector[1]
std_pf = np.sqrt(var_pf)

plt.plot(std_pf, mu_pf, '.') 
plt.show()
```

```python
# 1-(c) 3가지 stock이 추가된 (a) 케이스 (공매도 허용않음)
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) 
N = int(1e4)
mu_vector = np.array([0.15, 0.25, 0.3]) 
sigma_vector = np.array([0.1, 0.2, 0.15]) 
corr = 0.2

w1 = np.random.rand(N)
w2 = np.random.rand(N)*(1.0 - w1) 
w3 = 1.0 - w1 - w2 # 3가지 stock의 가중치 합은 1

mu_pf = w1*mu_vector[0] + w2*mu_vector[1] + w3*mu_vector[2]
var_pf = (w1**2)*(sigma_vector[0]**2) + (w2**2)*(sigma_vector[1]**2) + 2*w1*w2*corr*sigma_vector[0]*sigma_vector[1] \
+(w2**2)*(sigma_vector[1]**2) + 2*w2*w3*corr*sigma_vector[1]*sigma_vector[2] \
+(w3**2)*(sigma_vector[2]**2) + 2*w3*w1*corr*sigma_vector[2]*sigma_vector[0] # (ax+by_cz)^2 포맷

std_pf = np.sqrt(var_pf)
plt.plot(std_pf, mu_pf, '.') 
plt.show()
```
